```python
# Iterate through each row to remove stoprwords, punctuations, lowercase the text
import re
i = 0
for index, row in df.iterrows():
    review_text = str(row['review_text']).strip().lower()      
    review_tokens = review_text.split()
    
    result = []
    for token in review_tokens:
        if token not in stopwords_eng:
            result.append(re.sub(r'[^\w\s]','', token))
    df.at[index,'review_text'] = ' '.join(result)  
    df.at[index,'product_name'] = str(row['product_name']).strip().lower()
    df.at[index,'review_title'] = str(row['review_title']).strip().lower()
df  
```

```python
new_df = df[['product_name', 'product_type', 'review_text', 'rating', 'initial_sentiment']]
new_df
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

new_df['normal_sentiment_score'] = ''
for index, row in new_df.iterrows():    
    # Create a SentimentIntensityAnalyzer object
    analyzer = SentimentIntensityAnalyzer()
    # Use the analyzer to get a sentiment score for the text
    scores = analyzer.polarity_scores(row['review_text'])
    new_df.at[index,'normal_sentiment_score'] = scores['compound']

new_df
```

```python
# Calculating sentiment by 'normal_sentiment_score', -1 being most negative 0 being neutral and 1 being most positive

new_df['normal_sentiment'] = ''
new_df['rating_by_normal_sentiment'] = ''
for index, row in new_df.iterrows():
    if row['normal_sentiment_score'] < 0.0:
        if row['normal_sentiment_score'] <= -0.6:
            new_df.at[index,'normal_sentiment'] = 'negative'
            new_df.at[index,'rating_by_normal_sentiment'] = 1
        else:
            new_df.at[index,'normal_sentiment'] = 'negative'
            new_df.at[index,'rating_by_normal_sentiment'] = 2
    elif row['normal_sentiment_score'] == 0.0:
        new_df.at[index,'normal_sentiment'] = 'neutral'
        new_df.at[index,'rating_by_normal_sentiment'] = 3
    else:
        if row['normal_sentiment_score'] <= 0.5:
            new_df.at[index,'normal_sentiment'] = 'positive'
            new_df.at[index,'rating_by_normal_sentiment'] = 4
        else:
            new_df.at[index,'normal_sentiment'] = 'positive'
            new_df.at[index,'rating_by_normal_sentiment'] = 5
new_df
```

```python
# Excluding outliers or contradicting rating-review pairs example - where rating is excellent but review subitted is negative
for index, row in new_df.iterrows():
    if abs(int(row['rating_by_normal_sentiment'])-int(row['rating']))>=3:
        new_df.drop(index, inplace=True)
new_df
```
