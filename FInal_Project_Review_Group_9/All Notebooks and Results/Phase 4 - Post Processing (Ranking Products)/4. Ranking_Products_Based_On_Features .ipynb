```python
import re
# To get top 10 product_names for a given aspect and product_type
def get_top_products(aspect_name, product_type):
    # filter the DataFrame based on the product_type
    df_filtered = df[df['product_type'] == product_type]

    # create a list of tuples containing the product name and the aspect value
    product_aspect_list = []
    for index, row in df_filtered.iterrows():
        aspect_ratings = row['modified_aspects']
        if aspect_name in aspect_ratings:
            aspect_value = float(aspect_ratings[aspect_name])
            #print(type(aspect_value))
            product_aspect_list.append((row['product_name'], aspect_value))
    
    # sort the list in descending order by aspect value and return the top 10 product names
    sorted_product_aspect_list = sorted(product_aspect_list, key=lambda x: x[1], reverse=True)
    ranked_products = [x[0] for x in sorted_product_aspect_list[:10]]
    
    # apply the function to each element in the list
    ranked_products = [replace_multiple_spaces(text) for text in ranked_products]
    
    return ranked_products

# function to replace multiple whitespaces with single whitespace
def replace_multiple_spaces(text):
    return re.sub('\s+', ' ', text).strip()
```

```python
# Defining a static set of aspects for each category - 

book_aspects = ['plot/storyline','character development','writing style','theme','pacing','dialogue','setting','tone/mood','originality','ending','world-building','genre','narration','emotion','depth of research','structure/organization','clarity','suspense','symbolism','imagery', 'delivery']
dvd_aspects  = ["picture quality", "sound quality", "special features", "bonus content", "cast performances", "director's style", "cinematography", "screenplay", "plot twists", "character development", "action scenes", "comedy", "drama", "suspense", "romance", "science fiction", "fantasy", "historical accuracy", "costumes and set design", "adaptation from a book or other source material"]
electronics_aspects = ['performance/speed', 'battery life', 'display/screen quality', 'sound/audio quality', 'build quality/durability', 'design/aesthetics', 'ease of use', 'connectivity', 'software/operating system', 'storage capacity', 'camera/photography capabilities', 'gaming performance', 'energy efficiency', 'price/value for money', 'brand reputation', 'customer support/warranty', 'accessories included', 'security/privacy features', 'voice assistant integration', 'virtual/augmented reality compatibility', 'portability']
kitchen_houseware_aspects = ['material', 'quality', 'durability', 'design', 'ease of use', 'efficiency', 'capacity', 'safety', 'portability', 'brand reputation', 'price', 'customer support', 'accessories', 'environmentally friendly', 'performance', 'functionality']
```

```python
# Separate dataframe for top 10 ranked products in each category
books_ranked_products = pd.DataFrame()
dvd_ranked_products = pd.DataFrame()
electronics_ranked_products = pd.DataFrame()
kitchen_houseware_ranked_products = pd.DataFrame()
```

```python
for item in book_aspects:
    books_ranked_products[item] = get_top_products(item, 'books')
    
for item in dvd_aspects:
    dvd_ranked_products[item] = get_top_products(item, 'dvd')
    
for item in electronics_aspects:
    electronics_ranked_products[item] = get_top_products(item, 'electronics')
    
for item in kitchen_houseware_aspects:
    kitchen_houseware_ranked_products[item] = get_top_products(item, 'kitchen & housewares')
```
