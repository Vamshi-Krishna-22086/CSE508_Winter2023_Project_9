```python
# Remove two columns name is 'C' and 'D'
df.drop(['Unnamed: 0', 'index'], axis=1, inplace = True)
df
```

```python
type(df['aspect_ratings'][0])
```

```python
# Converting 'aspect_ratings' column from type string to dictionary 
# define a function to convert the string to a dictionary
import ast 
def str_to_dict(s):
    return ast.literal_eval(s)

# apply the function to the column
df['aspect_ratings'] = df['aspect_ratings'].apply(str_to_dict)
```

```python
type(df['aspect_ratings'][0])
```

```python
# use groupby and transform to add a new column with the count of rows in each group
import numpy as np
df['count'] = df.groupby('product_name')['product_name'].transform('count')

# replace NaN values with 0
df['count'] = df['count'].fillna(0)

# convert to integer
df['count'] = df['count'].astype(np.int64)
df
```

```python
# Merging similar aspects for similar products
from collections import defaultdict
aspects_dict = defaultdict(lambda: {'sum': 0, 'count': 0, 'products': set()})
for index, row in df.iterrows():
    for k, v in row['aspect_ratings'].items():
        aspects_dict[k]['sum'] += v
        aspects_dict[k]['count'] += 1
        aspects_dict[k]['products'].add(row['product_name'])

aspects_avg_dict = {}
for k, v in aspects_dict.items():
    if len(v['products']) == 1:
        aspects_avg_dict[k] = v['sum']
    else:
        aspects_avg_dict[k] = v['sum']/v['count']

df_grouped = df.groupby('product_name').agg({'count':lambda x: x.mode().iloc[0],'product_type': lambda x: x.mode().iloc[0], 'rating':'mean', 'normal_sentiment_score':'mean', 'rating_by_normal_sentiment':'mean', 'initial_sentiment':lambda x: x.mode().iloc[0], 'normal_sentiment':lambda x: x.mode().iloc[0], 'aspect_ratings': lambda x: {k:v for k,v in aspects_avg_dict.items() if k in x.values[0].keys()}})
# df_grouped = df.groupby('product_name').agg({'product_type': lambda x: x.mode().iloc[0], 
#                                             'rating':'mean', 
#                                             'normal_sentiment_score':'mean', 
#                                             'rating_by_normal_sentiment':'mean', 
#                                             'initial_sentiment':lambda x: x.mode().iloc[0], 
#                                             'normal_sentiment':lambda x: x.mode().iloc[0], 
#                                             'aspect_ratings': lambda x: {k:v for k,v in aspects_avg_dict.items() if k in x.values[0].keys()}})\
#                .reset_index()\
#                .assign(count_rows=lambda x: x.groupby('product_name')['product_name'].transform('count'))

df_grouped
```

```python
# reset index to create a new column with the old index values
df_grouped.reset_index(inplace=True)
df_grouped
```
