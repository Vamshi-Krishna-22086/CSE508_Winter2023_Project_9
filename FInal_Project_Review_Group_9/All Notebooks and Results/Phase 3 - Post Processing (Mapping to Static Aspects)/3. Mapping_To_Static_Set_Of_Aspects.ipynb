```python
def normalized_edit_distance(word1, word2): # Range 0 to 1 (0 - Most Similar)
    # calculate the edit distance using dynamic programming
    m, n = len(word1), len(word2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0:
                dp[i][j] = j
            elif j == 0:
                dp[i][j] = i
            elif word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(dp[i][j - 1], dp[i - 1][j], dp[i - 1][j - 1])
    
    # normalize the edit distance
    max_length = max(len(word1), len(word2))
    return dp[m][n] / max_length
```

```python
from nltk.corpus import wordnet

def wordnet_semantic_similarity(word1, word2): # Range 0 to 1 (1 Most Similar)
    synsets1 = wordnet.synsets(word1)
    synsets2 = wordnet.synsets(word2)
    max_sim = 0
    for syn1 in synsets1:
        for syn2 in synsets2:
            sim = syn1.path_similarity(syn2)
            if sim and sim > max_sim:
                max_sim = sim
    return max_sim
```

```python
from nltk.corpus import wordnet

def wupalmer_semantic_similarity(word1, word2): # Range 0 to 1, (1 is most similar)
    # get the synsets for each word
    try:
        synset1 = wordnet.synsets(word1)[0]
        synset2 = wordnet.synsets(word2)[0]
    except:
        return 0
    # calculate the similarity between the synsets using Wu-Palmer Similarity
    similarity = synset1.wup_similarity(synset2)
    
    return similarity
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm
import gensim.downloader as api

# Load the pre-trained GloVe model
model = api.load('glove-wiki-gigaword-300')
```

```python
def glove_semantic_similarity(word1, word2): # Range 0 to 1 (0 being most similar)
    # Get the word embeddings
    try:
        vec1 = model[word1]
        vec2 = model[word2]
    except KeyError:
        # Handle the case where one or both words are not in the vocabulary
        similarity = 0
    else:
        # Calculate the cosine similarity between the word embeddings
        similarity = dot(vec1, vec2) / (norm(vec1) * norm(vec2))
    return similarity
```

```python
# Converting 'aspect_ratings' column from type string to dictionary 
# define a function to convert the string to a dictionary
import ast 
def str_to_dict(s):
    return ast.literal_eval(s)

# apply the function to the column
df['aspect_ratings'] = df['aspect_ratings'].apply(str_to_dict)
```

```python
# Apply Semantic Similarity functions on aspects column 
# define a function to modify the dictionaries
import statistics

# Takes all retrieved aspects as input in a dictionary and maps them to static set of aspects initially defined by their syntactic and semantic similarity
def modify_dict(old_aspects, product_type):
    old_aspects_copy = old_aspects.copy()
    
    # define static aspects dependign upon the product type 
    if product_type == 'kitchen & housewares':
        static_aspects = ['material', 'quality', 'durability', 'design', 'ease', 'efficiency', 'capacity', 'safety', 'portability', 'brand', 'price', 'support', 'accessories', 'environment', 'performance', 'functionality']
    elif product_type == 'dvd':
        static_aspects = ["picture", "sound", "features", "bonus", "performance", "director", "cinematography", "screenplay", "twists", "character", "action", "comedy", "drama", "suspense", "romance", "fiction", "fantasy", "history", "costumes", "adaptation"]
    elif product_type == 'books':
        static_aspects = ['plot','character','writing','theme','pacing','dialogue','setting','tone','originality','ending','world-building','genre','narration','emotion','depth','structure','clarity','suspense','symbolism','imagery', 'delivery']
    elif product_type == 'electronics':
        static_aspects = ['performance', 'battery', 'display', 'sound', 'build', 'design', 'ease', 'connectivity', 'software', 'storage', 'camera', 'gaming', 'efficiency', 'price', 'brand', 'support', 'accessories', 'security', 'voice', 'reality', 'portability']
    else:
        return {}
    new_aspect_ratings = [[] for _ in range(len(static_aspects))]
    
    # Compare each old aspect with all static aspects for that product type and map to correct one
    for item in old_aspects_copy:
        true_item = item
        item = item.split()[0]
        item = item.split('_')[0]
        aspect_wise_similarities = [0 for _ in range(len(static_aspects))]

        max_similarity = 0
        if product_type in ['books', 'dvd', 'electronics', 'kitchen & housewares']:
            for i in range(len(static_aspects)):
                dist = normalized_edit_distance(static_aspects[i], item)
                
                # If aspect is syntactically closer to any of the static aspect, map it to that static aspect
                if dist < 0.5:
                    new_aspect_ratings[i].append(old_aspects_copy[true_item]) # Item work done
                    break
                else: # Else check for semantic similarity
                    item_similarities = []
                    item_similarities.append(1-dist)
                    item_similarities.append(wordnet_semantic_similarity(static_aspects[i], item))
                    item_similarities.append(wupalmer_semantic_similarity(static_aspects[i], item))
                    item_similarities.append(glove_semantic_similarity(static_aspects[i], item))

                    aspect_wise_similarities[i] = max(item_similarities)     
                    if i==len(static_aspects)-1:
                        # Considering max of all similarity techniques 
                        max_value = max(aspect_wise_similarities)
                        max_indices = [i for i, value in enumerate(aspect_wise_similarities) if value == max_value] 
                        
                        for index in max_indices:
                            new_aspect_ratings[index].append(old_aspects_copy[true_item])
            else:
                continue

                   
    # replace each list with its mean value, or 0 if it's empty
    new_aspect_ratings = [statistics.mean(sublist) if sublist else 0 for sublist in new_aspect_ratings]

    
    if product_type == 'kitchen & housewares':
        new_aspects_dict = dict(zip(kitchen_houseware_aspects, new_aspect_ratings))
    elif product_type == 'dvd':
        new_aspects_dict = dict(zip(dvd_aspects, new_aspect_ratings))
    elif product_type == 'books':
        new_aspects_dict = dict(zip(book_aspects, new_aspect_ratings))
    elif product_type == 'electronics':
        new_aspects_dict = dict(zip(electronics_aspects, new_aspect_ratings))

    return new_aspects_dict
```

```python
# apply the function to the column and store the output in a new column
df['modified_aspects'] = df.apply(lambda row: modify_dict(row['aspect_ratings'], row['product_type']), axis=1)
```
